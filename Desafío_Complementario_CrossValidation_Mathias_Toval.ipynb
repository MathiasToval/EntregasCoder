#Feature Engineering

```python
full= pd.concat([train,test], axis=0)
full.tail()
```

#Vuelo de larga distancia
Partiendo de que se considera un vuelo de larga distancia a aquellos que duran mas de 7 horas, se tiene en cuenta que la distancia aproximada es de unos 5000km. Por lo tanto se define una nueva variable que plantea si el vuelo es o no de larga distancia.

```python
full['distancia_vuelo'] = full['Flight Distance'].apply(lambda x: 0 if x < 5000 else 1)
```

#Rango etario del pasajero



Cuando consideramos en como de exigente es un cliente a la hora de clasificar un producto (en este caso, un vuelo de avión), podemos tener en cuenta el rango etario del mismo. Ya que una persona de la tercera edad suele tener mayores exigencias que un adolescente a la hora de viajar. Para esto se creará una nueva columna que define si la persona que viajó fue joven (menor de 18 años), adulto (entre 18 y 64 años) o adulto mayor (a partir de 64 años).

```python
#Defino una función para asignar cada grupo segun edad
def asignar_grupo_edad(edad):
    if edad < 18:
        return "Joven"
    elif edad < 65:
        return "Adulto"
    else:
        return "Adulto mayor"

full["Rango_etario"] = full["Age"].apply(lambda x: asignar_grupo_edad(x))
```

#Cantidad de vuelos realizados por el mismo pasajero

Además podemos tener en cuenta la cantidad de veces que tomó un vuelo un mismo pasajero para saber su exigencia según su experiencia.

```python
full['num_vuelos'] = full.groupby('id')['id'].transform('count')
```

#Demora total

Se creará una nueva columna que sume el tiempo de demora de salida y aterrizaje del mismo vuelo, para asi definir la demora total que haya experimentado un cliente y analizar como se relaciona con su satisfacción.

```python
full['Demora_total'] = full['Departure Delay in Minutes'] + full['Arrival Delay in Minutes']
```



---



# Encoding


Se aplica ONE HOT ENCODING a las variables categoricas relevantes.

```python
# Lista de columnas categóricas a codificar
columnas_categoricas = ['Gender','Customer Type','Type of Travel','Class','satisfaction', 'distancia_vuelo',"Rango_etario"]

# Se aplica one-hot encoding a las columnas categóricas pertinentes
dummies = pd.get_dummies(full[columnas_categoricas])

full_v2 = pd.concat([full, dummies], axis=1)

full_v2.drop(columnas_categoricas, axis=1, inplace=True)

full_v2
```

#Predicción

Se definen las variables X e Y para entrenar posteriormente. Nuestra feature objetivo será la variable "satisfaction_satisfied".

```python
X = full_v2.drop(['satisfaction_satisfied'], axis = 1)
Y = full_v2['satisfaction_satisfied'] # feature objetivo
```

Se crea la siguiente función que se utiliza para dividir un conjunto de datos en dos subconjuntos aleatorios llamados conjuntos de entrenamiento y prueba. Se utiliza para entrenar y evaluar modelos de aprendizaje automático.

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.4, random_state=42)
```

```python
#Normalizamos los datos
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Se aplica los modelos de predicción de Regresión Logistica, Arbol de decisiones y Random Forest

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#feature_names = X_train.columns

# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
logreg_importances = abs(logreg.coef_[0])

# Decision Tree
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred_dtree = dtree.predict(X_test)
dtree_importances = dtree.feature_importances_

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
rf_importances = rf.feature_importances_


# Evaluate the performance of each algorithm
def evaluate(y_test, y_pred):
    return {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1_score': f1_score(y_test, y_pred)
}

logreg_metrics = evaluate(y_test, y_pred_logreg)
dtree_metrics = evaluate(y_test, y_pred_dtree)
rf_metrics = evaluate(y_test, y_pred_rf)
#xgb_metrics = evaluate(y_test, y_pred_xgb)

# Compare the performance of each algorithm
print("Logistic Regression Metrics: ", logreg_metrics)
print("Decision Tree Metrics: ", dtree_metrics)
print("Random Forest Metrics: ", rf_metrics)


```

#PCA
A continuación aplicamos PCA a nuestros datos.

```python
#Volvemos a resetear las variables X e Y, utilizando el modelo de sprint seleccionado
X = full_v2.drop(['satisfaction_satisfied'], axis = 1) # feature matrix
Y = full_v2['satisfaction_satisfied'] # feature objetivo
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.4, random_state=42)
```

```python
#Normalizamos los datos para que PCA funcione mejor
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Aplicacion de PCA
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

Volvemos a aplicar los modelos

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
logreg_importances = abs(logreg.coef_[0])

# Decision Tree
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred_dtree = dtree.predict(X_test)
dtree_importances = dtree.feature_importances_

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
rf_importances = rf.feature_importances_


# Evaluate the performance of each algorithm
def evaluate(y_test, y_pred):
    return {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1_score': f1_score(y_test, y_pred)
}

logreg_metrics = evaluate(y_test, y_pred_logreg)
dtree_metrics = evaluate(y_test, y_pred_dtree)
rf_metrics = evaluate(y_test, y_pred_rf)


# Compare the performance of each algorithm
print("Logistic Regression Metrics: ", logreg_metrics)
print("Decision Tree Metrics: ", dtree_metrics)
print("Random Forest Metrics: ", rf_metrics)
```

# VALIDACIÓN CRUZADA
Se aplica tanto validación cruzada como optimización de hiperparametros al modelo con mejor rendimiento. En este caso, se plantea el Random Forest.

```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_classification

# Defino el modelo de Random Forest
rf_model = RandomForestClassifier()

# Aplico validación cruzada
scores = cross_val_score(rf_model, X, Y, cv=5)

# Calculo la precisión promedio y desviación estándar de los resultados
mean_accuracy = scores.mean()
std_accuracy = scores.std()

print("Precisión promedio:", mean_accuracy)
print("Desviación estándar de la precisión:", std_accuracy)
```
